# Initial Data Cut

Dataset is from https://www.kaggle.com/wordsforthewise/lending-club.
This initial download is 11 years of data. We're going to condense it down to 3 years which will still be a significant amount of data.  

```python
import pandas as pd
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)
```

These set option parameters are important because our dataframe is so large that pandas cuts off quite a bit of it. 

```python
#We turn the issue date column into a datetime so that we will be able to trim our data down to a certain time period. 
df = pd.read_csv('datasets/accepted_2007_to_2018Q4.csv', parse_dates=['issue_d'])
print(df.shape)
df.head()
```

Confirming that the issue date column is a datetime before we attempt to cut it. 

```python
print(df['issue_d'].dtype)
```

And here is where the actual cut occurs. This should give us every transaction from 2015-2017.

```python
#new dataframe is every row where the issue date is greater than 12/31/2014 and less than 01/01/2018
df2 = df[(df['issue_d'] > pd.to_datetime('2014-12-31')) & (df['issue_d'] < pd.to_datetime('2018-01-01'))]
```

```python
print(df2.shape)
df2.head()
```

If we did our job correctly, there should be only years from 2015-17

```python
df2['issue_d'].dt.year.value_counts()
```
